```python
#extract file fron zip file
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall()
zip_ref.close()
```

```python
#read data from the file and handle null values
import pandas as pd
df = pd.read_csv('orders.csv', na_values=['Not Available','unknown'])
df.head(20)
df['Ship Mode'].unique()
```

```python
rename columns names, make them lower case and replace with underscore
df.rename(columns={'Order Id':'order_id','City':'city'})
df.colums=df.columns.str.lower()
df.columns.str.lower()
df.columns=df.colums.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(5)
```

```python
#Derive new columns discount, sale price and profit
#df['discount']=df['list_price']*df['discount_percent']*.01
#df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df
```

```python
#convert order date from object data type to datetime
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
#drop cost price list and discount percent columns
df.drop(columns=['list_price','cost_price','discount_percent'], inplace=True)
```

```python
#load the data into sql server using replace option 
import sqlalchemy as sal
engine = sal.create_engine('mssql://Shresta/master?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn=engine.connect()
```

```python
#load the data into sql server using append option
df.to_sql('df_orders', con=conn , index=False, if_exists='append')
```
